# Introductory Fickean diffusion

## Fick's laws

**Fick's constitutive (first) law**, which is an expression for the diffusive flux of a chemical species, $J$ $[\mathrm{mol.m^{-2}}]$, as a function of a diffusion coefficient, $D$ $[\mathrm{m^{2}.s^{-1}}]$ and concentration, $c$ $[\mathrm{mol^{}.m^{-3}}]$. 

The nabla differential operator, $\nabla$, has inverse-length dimensions, i.e. its units are $[\mathrm{m^{-1}}]$. Hence, the term $\nabla c$ corresponds to a gradient of concentration of a chemical species.

$$\mathbf{J} = -D \nabla c$$

**Fick's second law**, is an statement of conservation of chemical species, i.e., the change of concentration with time (transient term, LHS) is equal to the divergence of the diffusive flux (diffusive term, RHS).

$$\frac{\partial c}{\partial t} = -\nabla \cdot \mathbf{J}$$

Substituting the expression for the diffusive flux vector, $\mathbf{J}$, the expression typically found in literature is obtained.

$$\frac{\partial c}{\partial t} = \nabla \cdot \left( D \nabla c \right)$$

## Finite volume discretization

To solve the diffusion equation numerically, we employ the finite volume method (FVM). In this derivation the departure-point will be conservation of chemical species equation.

The first step to establish a FVM scheme consists on integrating the governing equations over a control volume $\mathrm{P}$ with near-infinitesimal volume $V$, and over a near-infinitesimal time interval $\tau$.

### General form

**Integral equation:**

$$\int_{V}^{} \int_{\tau}^{} \frac{\partial c}{\partial t} dt dV = -\int_{\tau}^{} \int_{V}^{} \nabla \cdot \mathbf{J} dV dt$$

Gauss theorem is used to transform the divergence of diffusive flux into the total flux traversing the bounding surfaces of the control volume $\mathrm{P}$. Hence a volume integral is transformed into a surface integral.

$$\int_{V}^{} \int_{\tau}^{} \frac{\partial c}{\partial t} dt dV = -\int_{\tau}^{} \int_{A}^{} \mathbf{J} \cdot d\mathbf{A} dt$$

**General spatial and temporal integration**

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) V = - \left( \sum_{\mathrm{f}} \mathbf{J}_{\mathrm{f}} \cdot \mathbf{A}_{\mathrm{f}} \right) \tau $$

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = - \left( \sum_{\mathrm{f}} \mathbf{J}_{\mathrm{f}} \cdot \mathbf{A}_{\mathrm{f}} \right)$$

Here, $c_{\mathrm{P}}^{0}$ and $c_{\mathrm{P}}^{\tau}$ represent the concentration at the centroid of the control volume $\mathrm{P}$ at the beginning and the ending of the temporal integration interval, $\tau$.

The total flux through the bounding faces of the control volume, $\mathrm{V}$, is approximated as the sum of the normal flux through all the bounding faces.

### 1d discretization (interior cells)

#### Implementation #1: control volumes of arbitrary size

When a 1d geometry is considered, the control volume $\mathrm{P}$ has two neighbors, control volumes $\mathrm{W}$ and $\mathrm{E}$. Following from this, control volume $\mathrm{P}$ has two bounding faces, $\mathbf{A}_{\mathrm{w}}$ and $\mathbf{A}_{\mathrm{e}}$, referring to faces bounding in the western and eastern directions, respectively.

Then, the general equation takes the form:

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = - \mathbf{J}_{\mathrm{w}} \cdot \mathbf{A}_{\mathrm{w}} - \mathbf{J}_{\mathrm{e}} \cdot \mathbf{A}_{\mathrm{e}}$$

<div class="alert alert-info">

**NOTES:**

The western and eastern face bounding vectors, pointing in the western and eastern directions, are:

$$\mathbf{A}_{\mathrm{w}} = A_{\mathrm{w}} \left(-\mathbf{i} \right)$$

$$\mathbf{A}_{\mathrm{e}} = A_{\mathrm{e}} \left(+\mathbf{i} \right)$$

The species flux vectors, at the western and eastern faces, are:

$$\mathbf{J}_{\mathrm{w}} = - \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{w}} \left(-\mathbf{i} \right)$$

$$\mathbf{J}_{\mathrm{e}} = - \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{e}} \left(+\mathbf{i} \right)$$

A simple introductory approach is to use first-order approximations for the gradient terms:

$$\left. \frac{\partial c}{\partial x} \right|_{\mathrm{w}} \approx \frac{c_{\mathrm{W}} - c_{\mathrm{P}}}{\left|{\delta_{\mathrm{w}}}\right|}$$

$$\left. \frac{\partial c}{\partial x} \right|_{\mathrm{e}} \approx \frac{c_{\mathrm{E}} - c_{\mathrm{P}}}{\left|{\delta_{\mathrm{e}}}\right|}$$

where $c_{\mathrm{W}}$ and $c_{\mathrm{E}}$ are concentrations at the centroids of the neighboring control volumes, and $\delta_{\mathrm{w}}$ and $\delta_{\mathrm{e}}$ are absolute values of distances between the centroid of cell $\mathrm{P}$ and the centroids of neighboring control volumes.

</div>

Evaluating the dot-products yields:

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{w}} A_{\mathrm{w}} + \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{e}} A_{\mathrm{e}}$$

Using first-order approximations for the gradient terms:

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}} \right|} \left( c_{\mathrm{W}} - c_{\mathrm{P}}\right) + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}} \right|} \left( c_{\mathrm{E}} - c_{\mathrm{P}}\right)$$

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|} c_{\mathrm{W}} - \left( \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|} + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|} \right) c_{\mathrm{P}} + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|} c_{\mathrm{E}}$$

Note that $D_{\mathrm{w}}$ and $D_{\mathrm{e}}$ are effective diffusion coeffients at the western and eastern faces of the control volume $\mathrm{P}$, and that the evaluation of these coefficients is not arbitrary and for conservation (here the continuity of flux across the interface) to be respected one must take the harmonic mean of neighboring cells.

$$D_k = 2 \frac{D_i D_j}{D_i + D_j}$$

So far nothing has been said with respect to the instant at which the right-hand side is evaluated. The obvious solution would be to take $c^0$ as it is currently known, but for numerical stability reasons we will continue here with an implicit scheme, as follows:

$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|} c_{\mathrm{W}}^{\tau} - \left( \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|} + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|} \right) c_{\mathrm{P}}^{\tau} + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|} c_{\mathrm{E}}^{\tau}$$

Rearranging terms in a manner that is convenient (for later programming its solution):

$$-\left(\frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|}\right) c_{\mathrm{W}}^{\tau} + \left( \frac{V}{\tau} + \frac{D_{\mathrm{w}} A_{\mathrm{w}}}{\left| \delta_{\mathrm{w}}\right|} + \frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|} \right) c_{\mathrm{P}}^{\tau} -\left(\frac{D_{\mathrm{e}} A_{\mathrm{e}}}{\left| \delta_{\mathrm{e}}\right|}\right) c_{\mathrm{E}}^{\tau} = \left(\frac{V}{\tau} \right) c_{\mathrm{P}}^0$$

The resulting arrangement, can easily be coded into a tri-diagonal square matrix.

#### Implementartion #2: cartesian grid with regularly sized control volumes

The equations written above are general in the sense that they can be applied to any 1d coordinate system, e.g. the x-direction in orthogonal cartesian coordinates, or the r-direction on cylindrical or spherical coordinates.

When considering an orthogonal cartesian coordinate system, with regularly sized control volumes, some further simplifications can be done:

- The volume of a regular control volume on a cartesian orthogonal coordinate system, is given by: $V=\mathbf{\delta}_x \cdot (\mathbf{\delta}_y \times \mathbf{\delta}_z)$, where $\mathbf{\delta}_k$ is a vector giving the length of the control volume in the $x,y,z$ directions respectively.

- For instance, one can realize that when all the bounding face areas are of the same size $A = (\mathbf{\delta}_y \times \mathbf{\delta}_z)$ , i.e., $A = A_{\mathrm{w}}=A_{\mathrm{e}}$.

- Following from above, the distances between the two bounding faces of a control volume, are the same for all the control volumes, $\delta = \left| \mathbf{\delta}_x \right|$. Furthermore, the distances between cell centroids have this same value, i.e., $\delta = \left|\delta_{\mathrm{w}}\right|=\left|\delta_{\mathrm{e}}\right|$.

- Hence, the volume of a cell is calculated as, $V= \delta A$. On some references, a unit surface area is considered for the value of $A$, and the term is vanished. For the purpose of reminding that the FVM evaluates fluxes through bounding faces, the term will not be vanished hereafter.


Applying the assumptions stated above for a regular control volume on a cartesian orthogonal coordinate system:

$$-\left(\frac{D_{\mathrm{w}} A}{\delta}\right) c_{\mathrm{W}}^{\tau} + \left( \frac{\delta A}{\tau} + \frac{D_{\mathrm{w}} A}{\delta} + \frac{D_{\mathrm{e}} A}{\delta} \right) c_{\mathrm{P}}^{\tau} -\left(\frac{D_{\mathrm{e}} A}{\delta}\right) c_{\mathrm{E}}^{\tau} = \left(\frac{\delta A}{\tau} \right) c_{\mathrm{P}}^0$$

Multiplying the whole equation by $\delta$:

$$-\left(D_{\mathrm{w}} A \right) c_{\mathrm{W}}^{\tau} + \left(\frac{\delta^2 A}{\tau} + D_{\mathrm{w}} A + D_{\mathrm{e}} A\right) c_{\mathrm{P}}^{\tau} -\left(D_{\mathrm{e}} A \right) c_{\mathrm{E}}^{\tau} = \left(\frac{\delta^2 A}{\tau} \right) c_{\mathrm{P}}^{0}$$

Multiplying the whole equation by $\frac{\tau}{\delta^2 A}$ yields dimensionless coefficients:

$$-\left(\frac{D_{\mathrm{w}} \tau}{\delta^2} \right) c_{\mathrm{W}}^{\tau} + \left(1 + \frac{D_{\mathrm{w}} \tau}{\delta^2} + \frac{D_{\mathrm{e}} \tau}{\delta^2}  \right) c_{\mathrm{P}}^{\tau} - \left(\frac{D_{\mathrm{e}} \tau}{\delta^2} \right) c_{\mathrm{E}}^{\tau} = c_{\mathrm{P}}^{0}$$

$$-\alpha_{\mathrm{w}} c_{\mathrm{W}}^{\tau} + \left(1 + \alpha_{\mathrm{w}} + \alpha_{\mathrm{e}}  \right) c_{\mathrm{P}}^{\tau} - \alpha_{\mathrm{e}} c_{\mathrm{E}}^{\tau} = c_{\mathrm{P}}^{0}$$

### Boundary conditions (1d)

The departure equation is the same, but a special treatment is due in order to treat one of the fluxes as the boundary flux, $\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}}$.


$$\left(c_{\mathrm{P}}^{\tau} - c_{\mathrm{P}}^{0} \right) \frac{V}{\tau} = - \mathbf{J}_{\mathrm{w}} \cdot \mathbf{A}_{\mathrm{w}} - \mathbf{J}_{\mathrm{e}} \cdot \mathbf{A}_{\mathrm{e}}$$

#### Type 1:

Dirichlet boundary condition, fixed concentration at the boundary, i.e., $c_{\mathrm{B}} = c_{\mathrm{prescribed}}$.

Regarding the computational grid, we will make the choice of using ghost cells.
Assume the last cell on the east or west boundary is indexed as $\mathrm{P}$ and the ghost cell $\mathrm{G}$ to its western or eastern direction, i.e. fictious cell in place of the $\mathrm{W}$ or $\mathrm{E}$ cells. As the boundary composition being prescribed, we can propose a linear interpolation such as:

$$c_{\mathrm{B}} = \frac{c_{\mathrm{P}} + c_{\mathrm{G}}}{2} \Rightarrow c_{\mathrm{G}} = 2 c_{\mathrm{B}} - c_{\mathrm{P}}$$

The boundary flux is thus written as:

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{b}} A_{\mathrm{b}}$$

which is approximated as:

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \frac{D_{\mathrm{g}} A_{\mathrm{g}}}{\left| \delta_{\mathrm{g}} \right|} \left( c_{\mathrm{G}} - c_{\mathrm{P}}\right)$$

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \frac{D_{\mathrm{g}} A_{\mathrm{g}}}{\left| \delta_{\mathrm{g}} \right|} \left( 2 c_{\mathrm{B}} - c_{\mathrm{P}} - c_{\mathrm{P}}\right)$$

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \frac{D_{\mathrm{g}} A_{\mathrm{g}}}{\frac{1}{2} \left| \delta_{\mathrm{g}} \right|} \left( c_{\mathrm{B}} - c_{\mathrm{P}} \right)$$

The boundary flux $-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}}$ will substitute the expressions for $-\mathbf{J}_{\mathrm{w}} \cdot \mathbf{A}_{\mathrm{w}}$ or $-\mathbf{J}_{\mathrm{e}} \cdot \mathbf{A}_{\mathrm{e}}$ on the general equation, depending if the boundary is in the western or in the eastern direction of cell $\mathrm{P}$.

#### Type 2:

Neumann boundary condition, the flux of the chemical species is prescribed, $J_{\mathrm{B}} = J_{\mathrm{prescribed}}$.

The boundary flux is written as:

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = J_{\mathrm{B}} A_{\mathrm{B}}$$

The sign of $J_{\mathrm{B}}$ will indicate if the species is flowing in or out of the calculation domain, through the specified boundary.

The boundary flux $-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}}$ will substitute the expressions for $-\mathbf{J}_{\mathrm{w}} \cdot \mathbf{A}_{\mathrm{w}}$ or $-\mathbf{J}_{\mathrm{e}} \cdot \mathbf{A}_{\mathrm{e}}$ on the general equation, depending if the boundary is in the western or in the eastern direction of cell $\mathrm{P}$.

#### Type 3:

Robin boundary condition (with ghost-cell approach), the flux is proportional to the concentration difference between the boundary and a reference external composition $c_{\mathrm{\infty}}$.

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \left. D \frac{\partial c}{\partial x} \right|_{\mathrm{b}} A_{\mathrm{b}} \approx \frac{D_{\mathrm{g}} A_{\mathrm{g}}}{\frac{1}{2} \left| \delta_{\mathrm{g}} \right|} \left( c_{\mathrm{B}} - c_{\mathrm{P}} \right)$$

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = k_m \left( c_{\infty} - c_{\mathrm{B}} \right) A_{\mathrm{g}}$$

By equating the flux expressions, an equation for estimating the boundary concentration $c_{\mathrm{B}}$ (unknown a-priori) as a function of the concentration on cell $\mathrm{P}$ and the reference external composition $c_{\infty}$, is obtained.


$$\frac{D_{\mathrm{g}} A_{\mathrm{g}}}{\frac{1}{2} \left| \delta_{\mathrm{g}} \right|} \left( c_{\mathrm{B}} - c_{\mathrm{P}} \right) = k_m \left( c_{\infty} - c_{\mathrm{B}} \right) A_{\mathrm{g}}$$

$$c_{\mathrm{B}} = \frac{\displaystyle \frac{D_{\mathrm{g}}}{\frac{1}{2} \left| \delta_{\mathrm{g}} \right|} c_{\mathrm{P}} + k_m c_{\infty}}{\displaystyle \frac{D_{\mathrm{g}}}{\frac{1}{2} \left| \delta_{\mathrm{g}} \right|} + k_m}$$


A combinient expression for the boundary flux (for later programming) is obtained by assuming that a pseudo-steady flux profile prevails at the interface at all times (this ensures that continuity is preserved):

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = \frac{\left( c_{\infty} - c_{\mathrm{B}} \right) A_{\mathrm{g}}}{\displaystyle \frac{1}{k_m} + \frac{\frac{1}{2} \left|\delta_{\mathrm{g}}\right|}{D_{\mathrm{g}}}}$$

$$-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}} = k_{\mathrm{eq}}\left( c_{\infty} - c_{\mathrm{B}} \right) A_{\mathrm{g}}$$

$$k_{\mathrm{eq}}= \frac{\displaystyle \frac{1}{2}  \frac{k_m \left|\delta_{\mathrm{g}}\right|}{D_{\mathrm{g}}}}{\displaystyle k_m + \frac{1}{2} \frac{\left|\delta_{\mathrm{g}}\right|}{D_{\mathrm{g}}}}$$

The boundary flux $-\mathbf{J}_{\mathrm{b}} \cdot \mathbf{A}_{\mathrm{b}}$ will substitute the expressions for $-\mathbf{J}_{\mathrm{w}} \cdot \mathbf{A}_{\mathrm{w}}$ or $-\mathbf{J}_{\mathrm{e}} \cdot \mathbf{A}_{\mathrm{e}}$ on the general equation, depending if the boundary is in the western or in the eastern direction of cell $\mathrm{P}$.
